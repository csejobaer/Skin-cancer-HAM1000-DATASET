```python
import matplotlib.pyplot as plt
import numpy as np
def display_data_grid(images, labels, title):
    fig, axes = plt.subplots(4, 8, figsize=(15, 10))
    fig.suptitle(title, fontsize=16)

    for i, ax in enumerate(axes.flat):
        if i < len(images):
            ax.imshow(images[i].reshape(28, 28, 3))  # Reshape to image dimensions
            label_id = labels[i]
            label_name = classes[label_id][0]  # Get label name from 'classes' dictionary
            ax.set_title(label_name)
            ax.axis('off')
        else:
            ax.axis('off')

    plt.tight_layout()
    plt.show()

# Display a 4x8 grid of images and labels
display_data_grid(x[:32].values, y[:32].values, 'HAM10000 Dataset Images')
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Load the dataset (replace with your actual data loading code)
data = pd.read_csv('/content/drive/MyDrive/Machine-learning/hmnist_28_28_RGB.csv')

# Extract labels and image data
labels = data['label']
images = data.drop(columns=['label'])

# Create a dictionary to map labels to class names
classes = {
    4: 'Melanocytic nevi',
    6: 'Melanoma',
    2: 'Benign keratosis-like lesions',
    1: 'Basal cell carcinoma',
    5: 'Pyogenic granulomas and hemorrhage',
    0: 'Actinic keratoses and intraepithelial carcinomae',
    3: 'Dermatofibroma'
}

# Initialize a figure with subplots
fig, axes = plt.subplots(2, 4, figsize=(12, 6))
fig.suptitle('Single Image from Each Category', fontsize=16)

# Loop through each class and display a single image
for i, (class_id, class_name) in enumerate(classes.items()):
    # Find the index of the first image with the current class label
    idx = labels[labels == class_id].index[0]

    # Get the image data and reshape it to (28, 28, 3)
    image = images.iloc[idx].values.reshape(28, 28, 3)

    # Display the image with class name as title
    row = i // 4
    col = i % 4
    axes[row, col].imshow(image)
    axes[row, col].set_title(class_name, fontsize=10)
    axes[row, col].axis('off')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

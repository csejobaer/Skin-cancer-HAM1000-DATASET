```python
import matplotlib.pyplot as plt

# Data
model_names = [
    "Convolutional Neural Network",
    "Random Forest",
    "Naive Bayes",
    "K Nearest Neighbor (KNN)",
    "Support Vector Machine",
    "Decision Tree"
]

training_accuracy = [
    98.16767573356628,
    100.0,
    35.91136678385001,
    93.5336103121338,
    95.51507403856397,
    96.34601043997018
]

testing_accuracy = [
    98.17,
    98.77490145946521,
    36.0,
    94.0,
    96.0,
    96.0
]

# Create the figure and axes
fig, ax = plt.subplots(figsize=(10, 6))

# Creating the box plot
bp = ax.boxplot([training_accuracy, testing_accuracy], vert=True, patch_artist=True)

# Adding labels and title
ax.set_xticklabels(['Training Accuracy', 'Testing Accuracy'])
plt.ylabel('Accuracy (%)')
plt.title('Model Training and Testing Accuracy')

# Adding colors to the box plots
colors = ['b', 'g']
for box, color in zip(bp['boxes'], colors):
    box.set(facecolor=color)

# Adding a legend
ax.legend(bp['boxes'], ['Training Accuracy', 'Testing Accuracy'])

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
models = [
    "Convolutional Neural Network",
    "Random Forest",
    "Naive Bayes",
    "K Nearest Neighbor (KNN)",
    "Support Vector Machine",
    "Decision Tree",
    "GoogLeNet",
    "InceptionV3"
]

training_accuracy = [
    98.16767573356628,
    87.66849725411883,
    35.91136678385001,
    93.5336103121338,
    95.51507403856397,
    96.34601043997018,
    97.045,
    83
]

testing_accuracy = [
    98.17,
    70.7938092860709,
    36.0,
    94.0,
    96.0,
    96.0,
    97.045,
    80
]


# Create a 2x3 matrix of subplots
fig, axes = plt.subplots(3, 3, figsize=(15, 10))
fig.suptitle('Training and Testing Accuracy of Different Models', fontsize=16)

# Loop through models and create individual bar charts
for i, model in enumerate(models):
    row = i // 3
    col = i % 3
    ax = axes[row, col]

    # Plotting training and testing accuracy
    ax.bar(['Train', 'Test'], [training_accuracy[i], testing_accuracy[i]], color=['#6F1E51', '#0c2461'])
    ax.set_title(model)
    ax.set_ylim(0, 110)  # Setting y-axis limit to ensure the data fits

# Adjust spacing between subplots
plt.tight_layout()

# Display the plot
plt.show()
```

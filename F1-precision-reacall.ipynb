#inception

```python
import matplotlib.pyplot as plt

# Define the class labels
classes = ['AKIEC', 'BCC', 'BKL', 'DF', 'NV', 'VASC', 'MEL']
# Define the precision, recall, and F1-score values for each class
precision = [0.86, 0.74, 0.77, 0.94, 0.89, 0.99, 0.57]
recall = [0.80, 0.88, 0.50, 0.95, 0.64, 1.00, 0.84]
f1_score = [0.83, 0.80, 0.61, 0.95, 0.74, 0.99, 0.68]

# Create a bar plot
plt.figure(figsize=(10, 6))
plt.bar(classes, precision, label='Precision', width=0.2)
plt.bar(classes, recall, label='Recall', width=0.2, align='edge')
plt.bar(classes, f1_score, label='F1-Score', width=0.2, align='edge', alpha=0.6)

# Add labels and a legend
plt.xlabel('Classes')
plt.ylabel('Scores')
plt.title('Precision, Recall, and F1-Score for Each Class')
plt.legend()

# Show the plot
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

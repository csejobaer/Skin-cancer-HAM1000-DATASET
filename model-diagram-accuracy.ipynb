```python
import matplotlib.pyplot as plt

# Data
disease = ['AKIEC', 'BCC', 'BKL', 'DF', 'MEL', 'NV', 'VASC']
cnn_accuracy = [0.99, 0.99, 0.97, 0.99, 0.94, 0.99, 0.96]

# Create a bar chart
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
plt.bar(disease, cnn_accuracy, color='#0fb9b1')
plt.xlabel('Disease')
plt.ylabel('CNN Accuracy')
# plt.title('CNN Accuracy for Different Diseases')
plt.ylim(0.5, 1.0)  # Set the y-axis limits if needed
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

# Display the plot
plt.tight_layout()
plt.show()
```
